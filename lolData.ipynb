## What is the timespan of the dataset?

### SQL

```python
%%sql

    SELECT 
        MIN(game_creation) AS [Oldest],
        MAX(game_creation) AS [Newest]
    FROM data_match
```

### Django

```python
from django.db.models import Max, Min

query = Match.objects.all().aggregate(first=Min('game_creation'), last=Max('game_creation'))

last_date = datetime.fromtimestamp(query['last'] / 1000).strftime("%m/%d/%Y")
first_date = datetime.fromtimestamp(query['first'] / 1000).strftime("%m/%d/%Y")

show_queries()
```

### Answer

```python
print(f"Date Range from {first_date} to {last_date}")
```

## What is my highest winrate champion?

### SQL

```python
%%sql

SELECT
    champion_name,
    COUNT(win) filter (where win) as win_count,
    COUNT(match_id) as num_played,
    CAST((COUNT(win) filter (where win)) as FLOAT)
    / CAST((COUNT(match_id)) as FLOAT)
    as winrate
FROM data_participant
WHERE summoner_name = "Senor Verde"
GROUP BY champion_name
ORDER BY winrate DESC
```

### Django

```python
from data.models import Participant
from django.db.models import Count, IntegerField, FloatField
from django.db.models.functions import Cast

answer = Participant.objects.filter(
    summoner_name="Senor Verde"
).values(
    "champion_name"
).annotate(
    win_count=Sum(Cast("win", IntegerField())),
    total_games=Count("match_id"),
    winrate=Sum(Cast("win", FloatField()))/Cast(Count("match_id"), FloatField())
).order_by('-winrate')

show_queries()
```

```python
for item in answer:
    print(f"{item['champion_name']}: {item['win_count']} / {item['total_games']} ({item['winrate']})")
```

Let's look at champions where i've played more than 10 games

### SQL

```python
%%sql

SELECT
    champion_name,
    COUNT(win) filter (where win) as win_count,
    COUNT(match_id) as num_played,
    CAST((COUNT(win) filter (where win)) as FLOAT)
    / CAST((COUNT(match_id)) as FLOAT)
    as winrate
FROM data_participant
WHERE
    summoner_name = "Senor Verde"
GROUP BY champion_name
HAVING num_played > 10
ORDER BY winrate DESC
```

### Django

```python
reset_queries()

answer = Participant.objects.filter(
    summoner_name="Senor Verde"
).values(
    "champion_name"
).annotate(
    win_count=Sum(Cast("win", IntegerField())),
    total_games=Count("match_id"),
    winrate=Sum(Cast("win", FloatField()))/Cast(Count("match_id"), FloatField())
).filter(
    total_games__gte=10
).order_by('-winrate')

show_queries()
```

Looks like it doesn't run additional queries, since it has already ran the underlying query above
